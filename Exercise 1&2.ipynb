```python
import pandas as pd
import sqlite3
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
conn = sqlite3.connect("sleep_health.db")
df.to_sql("sleep_data", conn, if_exists="replace", index=False)

print("Data imported into SQL successfully!")
query = "SELECT * FROM sleep_data"

df_sql = pd.read_sql(query, conn)
print(df_sql.head())
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
gender_count = df["Gender"].value_counts()
gender_count.plot(kind="bar")
plt.title("Number of Participants by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
occupation_count = df["Occupation"].value_counts()
occupation_count.plot(kind="bar", figsize=(10,5))
plt.title("Number of Participants by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.show()
```
