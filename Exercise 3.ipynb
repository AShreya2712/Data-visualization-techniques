```python
import pandas as pd
df=pd.read_csv("C:/Users/DELL/Downloads/Sleep_health_and_lifestyle_dataset.csv")
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df)
```

```python
print(df.duplicated())
df.drop_duplicates(inplace=True)
print(df)
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Age'] < lower) | (df['Age'] > upper)]
print(outliers)
```

```python
Q1 = df['Sleep Duration'].quantile(0.25)
Q3 = df['Sleep Duration'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Sleep Duration'] < lower) | (df['Sleep Duration'] > upper)]
print(outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age']] = scaler.fit_transform(df[['Age']])
print(df[['Age']])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
print(df)
```
